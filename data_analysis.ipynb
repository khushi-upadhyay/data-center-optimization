```python
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv", low_memory=False, encoding='utf-8', nrows=3_170_000)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
missing = df.isnull().sum().sort_values(ascending = False)
print(missing[missing>0])
```

```python
df['fecha_servidor'] = pd.to_datetime(df['fecha_servidor'])
df.set_index('fecha_servidor', inplace=True)
```

```python
import matplotlib.pyplot as plt
df[['voltaje', 'corriente', 'potencia']].plot(figsize=(15, 5), alpha=0.7)
plt.title('Voltage, Current, Power over time')
plt.ylabel('Values')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
cols_to_plot = ['voltaje', 'corriente', 'potencia', 'frecuencia', 'energia']
df[cols_to_plot].hist(bins=50, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[['voltaje', 'corriente', 'potencia']])
plt.title("Boxplot of Electrical Features")
plt.show()
```

```python
df['weekday'] = df['weekday'].astype(str)
plt.figure(figsize=(10, 5))
sns.boxplot(x='weekday', y='energia', data=df)
plt.title("Energy consumption by weekday")
plt.xlabel("Weekday")
plt.ylabel("Energy (kWh)")
plt.show()
```

```python
df_daily = df['potencia'].resample('D').mean()
df_daily.plot(figsize=(15, 5), title="Daily Average Power Consumption")
plt.ylabel("Power (W)")
plt.show()
```
